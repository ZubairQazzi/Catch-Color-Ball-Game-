## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

data_url = (
    "https://raw.githubusercontent.com/"
    "ZubairQazzi/flyrank-ml-internship-zubair/"
    "main/data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(data_url)

total_pages = len(df)

declining_with_demand = (
    (df["trend_direction"] == "down")
    & (df["impressions_90d"] >= 100)
).sum()

high_visibility_pages = (
    df["impressions_90d"] >= 500
).sum()

print(f"Total pages in the starter dataset: {total_pages:,}")
print(f"Declining pages with at least 100 impressions: {declining_with_demand:,}")
print(f"Pages with at least 500 impressions: {high_visibility_pages:,}")
```

The starter dataset contains 30,000 pages. Among them, 13,152 pages are declining while still receiving at least 100 impressions, and 16,726 pages have at least 500 impressions. These numbers suggest there are many visible pages that may need prioritised review, making Refresh / Content Opportunity Scoring a useful lane to explore.
